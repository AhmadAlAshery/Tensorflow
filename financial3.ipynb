```python
sample_len = len(data[0])
train_len = int(np.ceil(sample_len*0.8)) 
test_len = sample_len - train_len
sample_len, train_len, test_len
```

```python
df_train = pd.DataFrame(data[0].sample(train_len, ignore_index=True)) 
df_test = pd.DataFrame(data[0].tail(test_len))
```

```python
scaler = MinMaxScaler()
```

```python
df_train[['Close']] = scaler.fit_transform(df_train[['Close']])
```

```python
df_train.tail()
```

```python
training_data = df_train[['Close']].values
```

```python
training_data.shape
```

```python
x_train = []
y_train = []
for i in range(60, len(training_data)):
    x_train.append(training_data[i-60:i, :])
    y_train.append(training_data[i,:])
```

```python
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
x_train.shape, y_train.shape
```

```python
model  = tf.keras.Sequential([
    tf.keras.layers.LSTM(100, return_sequences=True, input_shape=(60, 1)),
    tf.keras.layers.SimpleRNN(100, return_sequences=False),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(50, activation='relu'),
    tf.keras.layers.Dense(30, activation='relu'),
    tf.keras.layers.Dense(1)
])
```

```python
model.compile(loss=tf.keras.losses.Huber(),
              optimizer=tf.keras.optimizers.SGD(learning_rate=1e-7, momentum=0.9),
              metrics=["mae"])
```

```python
history = model.fit(x_train, y_train, epochs=100, callbacks=[tf.keras.callbacks.LearningRateScheduler(
    lambda epoch: 1e-7 * 10**(epoch / 30))])
```

```python
plt.semilogx(history.history["lr"], history.history["loss"])
plt.axis([1e-8, 1e-3, -0.25, 0.25])
```

```python
model.compile(loss=tf.keras.losses.Huber(),
              optimizer=tf.keras.optimizers.SGD(learning_rate=1e-4, momentum=0.9),
              metrics=["mae"])
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xval, ytrain, yval  = train_test_split(x_train, y_train, test_size=0.1, shuffle=True)
```

```python
xtrain.shape, xval.shape, ytrain.shape, yval.shape
```

```python
class ResetStatesCallback(tf.keras.callbacks.Callback):
    def on_epoch_begin(self, epoch, logs):
        self.model.reset_states()
```

```python
history_2 = model.fit(x_train, y_train, epochs=500, validation_data=(xval, yval), 
                      callbacks=[ tf.keras.callbacks.EarlyStopping(patience=50),
                                 tf.keras.callbacks.ModelCheckpoint("my_checkpoint.h5", save_best_only=True),
                                 ResetStatesCallback()
                                 ])
```

```python
history_2.history.keys()
```

```python
plt.plot(history_2.history['mae'], color='b', label="training mae")
plt.plot(history_2.history['val_mae'], color='r', label="val mae")
plt.legend()
```

```python
plt.plot(history_2.history['loss'], color='b', label="training loss")
plt.plot(history_2.history['val_loss'], color='r', label="val loss")
plt.legend()
```
